```python
import pandas as pd
from statsmodels.stats import weightstats as stests
```

```python
df = pd.read_csv('/content/2023_Countries by Population.csv')
df.head()
```

```python
df[['pop1980','pop2000']].describe()
```

```python
ztest ,pvalue = stests.ztest(df['pop1980'], x2=None, value=146)
print(float(pvalue))
print(ztest)
```

```python
if pvalue<0.05:
    print("Null hyphothesis rejected , Alternative hyphothesis accepted")
else:
    print("Null hyphothesis accepted , Alternative hyphothesis rejected")
```

```python
ztest,pval1=stests.ztest(df['pop1980'],df['pop2000'],value=100)
print(float(pval1))
print(ztest)
```

```python
if pval1<0.05:
    print("Null hyphothesis rejected , Alternative hyphothesis accepted")
else:
    print("Null hyphothesis accepted , Alternative hyphothesis rejected")
```
